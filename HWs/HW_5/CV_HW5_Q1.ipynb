```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the binary image
image_path = "/content/q1.png"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Threshold the image if not already binary
_, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# Find connected components
num_labels, labels_im = cv2.connectedComponents(binary_image)

# Map component labels to unique colors
label_hue = np.uint8(179 * labels_im / np.max(labels_im))
blank_ch = 255 * np.ones_like(label_hue)
colored_img = cv2.merge([label_hue, blank_ch, blank_ch])

# Convert to BGR for display
colored_img = cv2.cvtColor(colored_img, cv2.COLOR_HSV2BGR)
colored_img[label_hue == 0] = 0  # Set background label to black

# Display the image
plt.imshow(colored_img)
plt.title(f'Total number of connected components: {num_labels - 1}')
plt.show()
```
